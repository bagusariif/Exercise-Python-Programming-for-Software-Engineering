# Task 1. Calculate Average Sessions

**Task Descriptions**

- PacBook is a Social Media platform and they want to know what is their user Average Sessions and Session Duration for each data
- Given a three input for `load_time` (in minutes) and `exit_time` (in minutes)

- To calculate Session Duration, you can use this formula

$$
\text{Session}_\text{duration} = \text{exit}_\text{time} - \text{load}_\text{time}
$$

- To calculate Average Sessions, you can use this formula

$$
\text{Average}_\text{sessions} = \frac{\sum_{n=1}^{3}{\text{Session_duration}_{n}}}{\text{N_SESSIONS}}
$$

- But there are some constraints in this task
  - If the input for `load_time` and `exit_time` is less than zero or negative values, it will print the result **`"Not a Valid Input"`**

**Input - 1**

```py
load_time_1 = 5
exit_time_1 = 10

load_time_2 = 2
exit_time_2 = 15.5

load_time_3 = 1
exit_time_3 = 22.3
```

**Output - 1**

```py
Session 1 Duration (in minutes): 5
Session 2 Duration (in minutes): 13.5
Session 3 Duration (in minutes): 21.3
Average Session Duration (in minutes): 13.27
```

**Input - 2**

```py
load_time_1 = 0
exit_time_1 = 0

load_time_2 = 0
exit_time_2 = 0

load_time_3 = 0
exit_time_3 = 0
```

**Output - 2**

```py
Session 1 Duration (in minutes): 0
Session 2 Duration (in minutes): 0
Session 3 Duration (in minutes): 0
Average Session Duration (in minutes): 0
```

**Input - 3**

```py
load_time_1 = 3
exit_time_1 = 7

load_time_2 = -2
exit_time_2 = 5

load_time_3 = 3
exit_time_3 = 156
```

**Output - 3**

```py
Not a Valid Input
```

````python
# Write your code here
load_time_1 = 3
exit_time_1 = 7

load_time_2 = -2
exit_time_2 = 5

load_time_3 = 3
exit_time_3 = 156

#formula
ses_1 = exit_time_1 - load_time_1
ses_2 = exit_time_2 - load_time_2
ses_3 = exit_time_3 - load_time_3

if load_time_1 < 0 or exit_time_1 < 0:
  print ("Not a Valid Input")
else:
  print(f"Session 1 Duration (in minutes): {ses_1}")

if load_time_2 < 0 or exit_time_2 < 0:
    print ("Not a Valid Input")
else:
    print(f"Session 2 Duration (in minutes): {ses_2}")

if load_time_3 < 0 or exit_time_3 < 0:
  print ("Not a Valid Input")
else:
  print(f"Session 3 Duration (in minutes): {ses_3}")

#average
avr = (ses_1 + ses_2 + ses_3) / 3
print(f"Average Session Duration (in minutes): {round(avr, 2)}")
````

- NIK is not some random numbers. You can extract information from the NIK.
  - The first six digits are a province, regency, and district code
  - The next six digits are the date of birth; DD-MM-YY, respectively
    - For men, the date of birth ranged between 01-31
    - For women, the date of birth is added with 40, hence ranged between 41-71.
  - The last 4 digits are a serial number of NIK issuance.


- Given **two** NIK, your tasks are
  1. Extract the date of birth of each NIK
  2. Extract their gender.
  3. Extract whether they have similar gender.
  4. Extract whether they have similar month of birth.

````python
# Write your code here
nik_1 = 1571011709860003
nik_2 = 2404075109910009

# Extract 7-12 digits to classified the Date of Birth from NIK
nik_1_birth_date = str(nik_1)[6:12]
nik_2_birth_date = str(nik_2)[6:12]

# Simplify the date by substract 400000
if int(nik_1_birth_date) > 310000:
    nik_1_birth_date = str(int(nik_1_birth_date) - 400000)

if int(nik_2_birth_date) > 310000:
    nik_2_birth_date = str(int(nik_2_birth_date) - 400000)

# Formatted the date with "-" for each 2 digits to create DD-MM-YY
formatted_nik_1_birth_date = (
    nik_1_birth_date[0:2] + "-" +
    nik_1_birth_date[2:4] + "-" +
    nik_1_birth_date[4:6]
)

formatted_nik_2_birth_date = (
    nik_2_birth_date[0:2] + "-" +
    nik_2_birth_date[2:4] + "-" +
    nik_2_birth_date[4:6]
)

# Print of Date of Birth from NIK
print(f"Date of Birth")
print(f" - NIK 1 : {formatted_nik_1_birth_date}")
print(f" - NIK 2 : {formatted_nik_2_birth_date}")
print("\n")

# Extract 7-8 digits to classified date (DD) from both NIK and create gender variables
date_nik_1 = int(str(nik_1)[6:8])
date_nik_2 = int(str(nik_2)[6:8])
gender_1 = "Male"
gender_2 = "Female"

# Identify Gender by date from NIK
if 0 <= date_nik_1 <= 31:
    gender_nik_1 = gender_1
elif 41 <= date_nik_1 <= 71:
    gender_nik_1 = gender_2
else:
     gender_nik_1 = "Date range must be within 01-31 and 41-71"

if 0 <= date_nik_2 <= 31:
    gender_nik_2 = gender_1
elif 41 <= date_nik_2 <= 71:
    gender_nik_2 = gender_2
else:
    gender_nik_2 = "Date range must be within 01-31 and 41-71"

# Print Gender based on date
print(f"Gender")
print(f"NIK 1: {gender_nik_1}")
print(f"NIK 2: {gender_nik_2}")
print("\n")

# Check the similarity of gender
print("Do they have similar gender?")
if gender_nik_1 == gender_nik_2:
    print("Yes.")
else:
    print("No.")
print("\n")

# Extract month from date of birth
month_nik_1 = nik_1_birth_date[2:4]
month_nik_2 = nik_2_birth_date[2:4]

# Create dictionary for month
month_of_year = {
    "01" : "January",
    "02" : "February",
    "03" : "March",
    "04" : "April",
    "05" : "May",
    "06" : "June",
    "07" : "July",
    "08" : "August",
    "09" : "September",
    "10" : "October",
    "11" : "November",
    "12" : "December"

}

# Create variable (month_nik) to collect keys (month_nik_1 and month_nik_2) from dict (month_of_year)
month_nik = month_of_year.get(month_nik_1 and month_nik_2)

# Check similarity of month of birth
print("Do they have similar month of birth?")
if month_nik_1 == month_nik_2:
    print(f"Yes, it is on {month_nik}.")
else:
    print("No.")
````

# Task 4. Calculate Pajak Penghasilan (Simplified)

Category: Hard

---

**Task Descriptions**
- According to Pemotongan Pajak Penghasilan PPh Pasal 21 Tahun 2016, we can calculate our income tax as (we'll write in Bahasa Indonesia for avoiding missinterpretation)

- **Penghasilan tidak kena pajak (PTKP)**
  - Penghasilan Rp 54.000.000 setahun untuk wajib pajak.
  - Tambahan Rp 4.500.000 diberikan jika wajib pajak menikah.
  - Tambahan Rp 4.500.000 lagi jika punya 1 orang anak atau Rp 9.000.000 jika punya 2 orang anak. Apabila jumlah tanggungan lebih dari itu, maka tetap dihitung 3 orang tanggungan.

- **Tarif Pajak Penghasilan**
  - 5% untuk penghasilan kena pajak (PKP) sampai dengan 60 Juta
  - 15% untuk PKP di atas 60 juta sampai dengan 250 Juta
  - 25% untuk PKP di atas 250 juta sampai dengan 500 juta
  - 30% untuk PKP di atas 500 juta sampai dengan 5 miliar
  - 35% untuk PKP di atas 5 miliar

- Your task is calculate the income tax given
  - Monthly income
  - Marital status
  - The number of dependents (jumlah tanggungan)

**Input - 1**

```python
monthly_income = 4750000
marital_status = 'single'
n_dependents = 0
```

**Output - 1**
```
Penghasilan neto setahun:
57000000

PTKP TK/0:
-54000000

Penghasilan Kena Pajak:
3000000

PPh 21 terutang setahun:
150000
```

**Input - 2**

```python
monthly_income = 15000000
marital_status = 'married'
n_dependents = 3
```

**Output - 2**


````python
# Write your code here
monthly_income = 15000000
marital_status = 'married'
n_dependents = 3

# Penghasilan neto setahun
yearly_neto = monthly_income * 12

# Penghasilan tidak kena pajak (PTKP)
ptkp = 54000000

# Adjusting PTKP based on marital status and dependents.
if marital_status == "married":
    ptkp += 4500000  # Additional Rp 4.500.000 if married.
    if n_dependents == 1:
        ptkp += 4500000  # Additional Rp 4.500.000 if you have 1 dependents.
    elif n_dependents == 2:
        ptkp += 9000000  # Additional Rp 9.000.000 if you have 2 dependents.
    elif n_dependents >= 3:
        ptkp += 13500000 # Additional Rp 13.500.000 if you have 3 dependents or more.

# Penghasilan kena pajak (PKP)
pkp = yearly_neto - ptkp

# PPh 21
if pkp > 0:
    if 0 <= pkp <= 60000000:  # 5% tax for PKP until 60 Million.
        pph = pkp * 0.05
    elif 60000000 <= pkp <= 250000000:  # 15% tax for PKP above 60 Million until 250 Million.
        pph = pkp * 0.15
    elif 250000000 <= pkp <= 500000000:  # 25% tax for PKP above 250 Million until 500 Million.
        pph = pkp * 0.25
    elif 500000000 <= pkp <= 5000000000:  # 30% tax for PKP above 500 Million until 5 Billion.
        pph = pkp * 0.30
    else:  # 35% tax for PKP above 5 Billion.
        pph = pkp * 0.35

print("Penghasilan neto setahun:")
print(yearly_neto)
print("\nPTKP TK/3:")
print(-ptkp)
print("\nPenghasilan Kena Pajak:")
print(pkp)
print("\nPPh 21 terutang setahun:")
print(pph)
````
